```python
#now lets backtest 10, 30, 60 day rate of changes for a bunch of FX pairs

#pairs will come from yfinance

# we will nee eurchf, Eurchf, chfbrl, chfjpy, chfpln, chfcad, usdchf, chfmxn

pairs = ['EURCHF=X', 'EURCHF=X', 'CHFPLN=X', 'CHFJPY=X', 'CHFCAD=X', 'USDCHF=X', 'CHFMXN=X']

# lets get the data for the pairs

import matplotlib
width, height = 20, 10 # example with 15x6 inches
matplotlib.rcParams['figure.figsize'] = [width, height]

import yfinance as yf

for i in pairs:

    print ("**********" + i + "**********")

    df = yf.download(i, start="1950-03-01", end="2023-09-15")
    df['ROC_10D'] = df['Close'].pct_change(periods=10)
    df['ROC_30D'] = df['Close'].pct_change(periods=30)
    df['ROC_60D'] = df['Close'].pct_change(periods=60)
    df['30 Day Forward Return'] = df['Close'].pct_change(periods=-30)
    df = df.dropna()


    # now lets make a regression which will have the rate of change as the independent variable and the dependent variable and 30 day forward return as the dependent variable

    import statsmodels.api as sm

    # we will use the not OLS model to make a regression model

    model = sm.OLS(df['30 Day Forward Return'], df['ROC_30D']).fit()

    # lets get the summary of the model

    print(model.summary())

    # now plot the regression line

    import matplotlib.pyplot as plt

    plt.scatter(df['ROC_30D'], df['Close'].pct_change(periods=-30))
    plt.plot(df['ROC_30D'], model.predict(df['ROC_30D']), color='red')
    plt.xlabel('ROC_30D')
    plt.ylabel('% Change 30 days later')

    # draw the regression line

    plt.plot(df['ROC_30D'], model.predict(df['ROC_30D']), color='red')

    
    # add title to the plot

    plt.title('ROC_30D VS % Change 30 days later for ' + i)

    plt.show()

```
